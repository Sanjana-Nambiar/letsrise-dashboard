```python
decision_to_consequence = consequence_df.set_index('Decision Statement')['Consequence'].to_dict()
decision_to_consequence
```

```python
# Function to clean text by removing special characters and converting to lowercase
def clean_text(text):
    return re.sub(r'\W+', ' ', text).lower().strip()

# Clean the decision statements in consequence_df and create a mapping
consequence_df['Cleaned Decision Statement'] = consequence_df['Decision Statement'].apply(clean_text)
decision_to_consequence = consequence_df.set_index('Cleaned Decision Statement')['Consequence'].to_dict()
decision_to_consequence
```

```python
# Initialize a dictionary to store the scores for each user
user_scores = {}

# Iterate through each user in the assessment dataframe
for idx, user in assessment_df.iterrows():
    user_id = user['User_id']
    user_scores[user_id] = {consequence: 0 for consequence in consequence_df['Consequence'].unique()}
    
    # Iterate through the 45 questions and update scores based on the user's responses
    for q in range(1, 46):
        response_statement = user[f'Question {q}']
        cleaned_response = clean_text(response_statement)
        if cleaned_response in decision_to_consequence:
            consequence = decision_to_consequence[cleaned_response]
            user_scores[user_id][consequence] += 1  # Increment the score for the relevant consequence

# Convert the user_scores dictionary to a DataFrame for better analysis
scores_df = pd.DataFrame.from_dict(user_scores, orient='index')
scores_df.head(5)
```

```python
# Initialize a dictionary to store the scores for each user
user_scores = {}
user_info = {}  # Dictionary to store user information

# Iterate through each user in the assessment dataframe
for idx, user in assessment_df.iterrows():
    user_id = user['User_id']
    username = user['Name']  # Assuming you have a column for username
    user_scores[user_id] = {consequence: 0 for consequence in consequence_df['Consequence'].unique()}
    user_info[user_id] = username

    # Iterate through the 45 questions and update scores based on the user's responses
    for q in range(1, 46):
        response_statement = user[f'Question {q}']
        cleaned_response = clean_text(response_statement)
        if cleaned_response in decision_to_consequence:
            consequence = decision_to_consequence[cleaned_response]
            user_scores[user_id][consequence] += 1  # Increment the score for the relevant consequence

# Convert the user_scores dictionary to a DataFrame for better analysis
scores_df = pd.DataFrame.from_dict(user_scores, orient='index')

# Add a new column for the username
scores_df['Username'] = scores_df.index.map(user_info)

# Reorder columns to have 'User_id' and 'Username' at the beginning
scores_df.reset_index(inplace=True)
scores_df.rename(columns={'index': 'User_id'}, inplace=True)
scores_df = scores_df[['User_id', 'Username'] + [col for col in scores_df.columns if col not in ['User_id', 'Username']]]

scores_df.head(5)
```

```python
data_model = retrieve_data("1rUDTXpzyyLuHXc_Tq8lANaqxbyaOR6DOLAP4Nq7X9Q0", "Assessment Form")
data_model_df = pd.DataFrame(data_model)
data_model_df['Cleaned Decision Statement'] = data_model_df['Decision Statement'].apply(clean_text)
data_model_df.head(5)
```

```python
consequence_set = set(consequence_df['Cleaned Decision Statement'])
data_model_set = set(data_model_df['Cleaned Decision Statement'])

# Check how many cleaned decision statements in data_model_df are present in consequence_df
data_model_df['Exists in Consequence'] = data_model_df['Cleaned Decision Statement'].apply(lambda x: x in consequence_set)

# Calculate the number of matches
num_matches = data_model_df['Exists in Consequence'].sum()
total_decision_statements = len(data_model_df)
percentage_matches = (num_matches / total_decision_statements) * 100

# Output the results
print(f"Total decision statements in data_model_df: {total_decision_statements}")
print(f"Number of matching decision statements: {num_matches}")
print(f"Percentage of matching decision statements: {percentage_matches:.2f}%")

# List the statements that are missing or not matching
non_matching_statements = data_model_df[~data_model_df['Exists in Consequence']]
non_matching_statements.to_csv('non_matching_statements.csv', index=False)

# Find statements that are in consequence_df but not in data_model_df
extra_consequence_statements = consequence_df[~consequence_df['Cleaned Decision Statement'].isin(data_model_set)]
extra_consequence_statements.to_csv('extra_consequence_statements.csv', index=False)

```
